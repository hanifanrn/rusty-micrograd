```rust
fn f(x: f64) -> f64 {
    return 3.0 * x.powi(2) - 4.0 * x + 5.0;
}
```

```rust
f(3.0)
```

```rust
:dep plotters = { version = "^0.3.5", default_features = false, features = ["evcxr", "all_series", "all_elements"] }
extern crate plotters;
use plotters::prelude::*;

let mut num = -5.0;
let mut xs: Vec<f64> = vec![];
while num <= 5.0 {
    xs.push(num);
    num += 0.25;
}

let figure = evcxr_figure((640, 480), |root| {
    root.fill(&WHITE)?;
    let mut chart = ChartBuilder::on(&root)
        .caption("f(x) = 3.0 * x.powi(2) - 4.0 * x + 5.0", ("Arial", 50).into_font())
        .margin(5)
        .x_label_area_size(30)
        .y_label_area_size(30)
        .build_cartesian_2d(-5f64..5f64, -0.1f64..100f64)?;

    chart.configure_mesh().draw()?;

    chart.draw_series(LineSeries::new(
        xs.into_iter().map(|x| (x, 3.0 * x.powi(2) - 4.0 * x + 5.0)),
        &RED,
    )).unwrap()
        .label("y = 3.0 * x.powi(2) - 4.0 * x + 5.0")
        .legend(|(x,y)| PathElement::new(vec![(x, y), (x + 20, y)], &RED));

    chart.configure_series_labels()
        .background_style(&WHITE.mix(0.8))
        .border_style(&BLACK)
        .draw()?;
    Ok(())
});
figure
```

```rust
let h = 0.001;

// Input
let mut a = 2.0;
let mut b = -3.0;
let c = 10.0;

let d1 = a * b + c;
b += h;
let d2 = a * b + c;

println!("d1: {}", d1);
println!("d2: {}", d2);
println!("slope: {}", (d2 - d1)/h);
```

```rust
#[derive(Debug)]
struct Value {
    data: f64,
}

impl Value {
    fn new(data: f64) -> Self {
        Value { data }
    }

    fn add(&self, other: &Value) -> Self {
        let out = self.data + other.data;
        return Value { data: out };
    }

    fn mul(&self, other: &Value) -> Self {
        let out = self.data * other.data;
        return Value { data: out };
    }

}
```

```rust
let a = Value::new(2.0);
let b = Value::new(-3.0);
let c = Value::new(10.0);
```

```rust
println!("{:?}", a);
```

```rust
println!("{:?}", a.add(&b));
```

```rust
(a.mul(&b)).add(&c)
```
